```python
import pyLDAvis
import pickle
import pyLDAvis.gensim as gensimvis
```

```python
# load full LDA model
fileObject = open('./models/model_full_lda.sav','rb')  
model_full = pickle.load(fileObject)

# load full LDA corpus
fileObject = open('./lda_topics/corpus_full_lda.sav','rb')  
corpus = pickle.load(fileObject)

# load full LDA dictionary
fileObject = open('./lda_topics/dictionary_full_lda.sav','rb')  
dictionary = pickle.load(fileObject)
```

```python
# load real_news LDA model
fileObject = open('./models/lda_model_real.sav','rb')  
model_real = pickle.load(fileObject)

# load real_news LDA corpus
fileObject = open('./lda_topics/corpus_real_lda.sav','rb')  
corpus_real = pickle.load(fileObject)

# load real_news LDA dictionary
fileObject = open('./lda_topics/dictionary_real_lda.sav','rb')  
dictionary_real = pickle.load(fileObject)
```

```python
# load fake_news LDA model
fileObject = open('./models/lda_model_fake.sav','rb')  
model_fake = pickle.load(fileObject)

# load fake_news LDA corpus
fileObject = open('./lda_topics/corpus_fake_lda.sav','rb')  
corpus_fake = pickle.load(fileObject)

# load fake_news LDA dictionary
fileObject = open('./lda_topics/dictionary_fake_lda.sav','rb')  
dictionary_fake = pickle.load(fileObject)
```

```python
vis_data = gensimvis.prepare(model_full, corpus, dictionary)
```

## LDA Visualization: All Articles

```python
pyLDAvis.display(vis_data)
```
